```python
outDataCardsDir = '/uscms/home/amalbert/nobackup/el9_work/CMSSW_14_1_0_pre4/src/HiggsAnalysis/CombinedLimit/Datacards/first_ABCD_counts/'
signal_name = 'HNL_mu'
mass = 2
sig_unc = {'dummy':[[0.2,0.2,0.2,0.2]]}
sig_unc_name = ['dummy']
signal_region='A'
prefix='bin'


ctaus = [10, 25, 50, 100, 250, 500, 750, 1000, 2500, 5000, 7500, 10000]
for ctau in ctaus:
    print(f"running for ctau {ctau}")
    cutflow_signal,ABCD_signal = Produce_Cutflow_v2.makeCutflow(signal_events, cfg_file='cuts_sensitivity_mu_DNN.yaml', reweight_ctau=ctau, isMC=True, ABCD=True, blind=False, flavor="Mu")
    #print(ABCD_signal)
    #ctau = 1000

    list_signal = [float(counts[:counts.find("+-")]) for counts in ABCD_signal['Counts']]
    list_data = [float(counts[:counts.find("+-")]) for counts in ABCD_data['Counts']]

    signal_rate = {f'{signal_name}':[list_signal[3],list_signal[1],list_signal[0], list_signal[2]]}
    norm = 1/(signal_rate[f'{signal_name}'][0])
    bkg_rate = [list_data[3],list_data[1],list_data[0], list_data[2]]
    observation = bkg_rate
    #print(signal_rate)
    #print(norm)
    analysis_helpers.make_datacard_2tag(outDataCardsDir,signal_name,  signal_rate, norm, bkg_rate, observation, sig_unc, sig_unc_name, signal_region, prefix, mass, ctau)
    
```
